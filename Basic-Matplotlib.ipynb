# Matplotlib

```python
#importing matplotlib.pyplot library and numpy

import matplotlib.pyplot as plt
import numpy as np
```

```python
data=np.arange(10)
data
```

# A basic visualization

```python
plt.plot(data)
```

```python
# Lets plot by defining x and y values

x=np.arange(10)
y=x**2

plt.plot(x,y)
```

```python
# adding title and axis label

plt.title("y=x**2 graph") #title
plt.xlabel("x- axis") #x-axis label
plt.ylabel("y-axis") #y-axis label
plt.plot(x,y) 
```

```python
# changing line width, color and marker

plt.plot(x,y,color='blue',linewidth=2,marker='*',markersize=10)
```

Ploting two graphs in a single plot

```python
plt.plot(x,y,'r',x,y**2,'b')
```

# Creating Figures and subplots

```python
fig=plt.figure()
```

Creating subplots - (2,2,1) - In this 2,2 denotes 2x2=4 subplots and 1 denotes the first subplot.

```python
fig=plt.figure()
plot_1=fig.add_subplot(2,2,1)
plot_2=fig.add_subplot(2,2,2)
plot_3=fig.add_subplot(2,2,3)
plot_4=fig.add_subplot(2,2,4)
```

```python
# first we need to define the figure
fig=plt.figure(figsize=(10,10))

#start define about the subplots separately

#subplot-1
plot_1=fig.add_subplot(2,2,1)
plot_1.plot(x,y,color='red')
plot_1.set_title('x vs y')

#subplot-2
plot_2=fig.add_subplot(2,2,2)
plot_2.plot(x,y*2,'b')  # we can also define color as like this too
plot_2.set_title('x vs 2y')

#subplot-3
plot_3=fig.add_subplot(2,2,3)
plot_3.plot(x,y,'y*')
plot_3.set_title('x vs y with marker')

#subplot-4
plot_4=fig.add_subplot(2,2,4)
plot_4.plot(x,y*2,'b*')
plot_4.set_title('x vs 2y with markers')

plt.suptitle('2x2 subplots',fontsize=20) # adding fontsize
plt.show()
```

# Adding legends

```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)

line_1 = np.arange(0.0, 5.0, 0.1)
line_2 = np.arange(0.0, 5.0, 0.02)
line_3 = np.arange(0.0, 2.0, 0.01)

#labels of respective plots should be defined inside label arugment

ax.plot(line_1**2, label='line-1') 
ax.plot(line_2**2, label='line-2')
ax.plot(line_3**3, label='line-3')
ax.legend()
plt.show()
```
